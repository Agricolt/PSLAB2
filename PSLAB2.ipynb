```python
from sympy import *
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

p = 1 + 0.41
T=2
w=np.pi

def f(t):
  return p*t**2

def fa1(t):
  return f(t)*np.cos(w*t)

def fa2(t):
  return f(t)*np.cos(2*w*t)

a0=2/T*quad(f, -1, 1)[0]
a1=2/T*quad(fa1, -1, 1)[0]
a2=2/T*quad(fa2, -1, 1)[0]

def periodically_continued(a, b):
    interval = b - a
    return lambda f: lambda x: f((x - a) % interval + a)

y = periodically_continued(-1, 1)(lambda x: p*x**2)

def F(t):
  return (a0/2)+a1*np.cos(np.pi*t)+a2*np.cos(2*np.pi*t)
x = np.linspace(-1, 1, 300)
t = np.linspace(-3, 3, 300)
fig, ax=plt.subplots()
ax.plot(t, F(t))
ax.plot(t, y(t))
ax.grid(True)
plt.show()
```
